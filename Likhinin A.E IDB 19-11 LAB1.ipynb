#### № 2.4

```python
def transformation_plot(points, matrix):
    figure,graph = plt.subplots(1, 2, figsize = (20, 10))
    
    x = points[:, 0]
    y = points[:, 1]
    sum = x*y
    
    graph[0].scatter(x, y, 100, sum)
    trans_points = numpy.dot(points, matrix)
    trans_x = trans_points[:, 0]
    trans_y = trans_points[:, 1]
    graph[1].scatter(trans_x, trans_y, 50, sum)
    
transformation_plot(numpy.random.random (size = (200,2)),numpy.array([[16,4],[30,3]]))
```

#### № 2.6

```python
v = numpy.random.random(size=(55,2))-0.5
q = numpy.array([10, 1])


count1 = 0
count2 = 0

for i in v:
    cos = i.dot(q)/(numpy.linalg.norm(q)*numpy.linalg.norm(i))
    if numpy.arccos(cos)*180/numpy.pi < 90:
        count1 = count1 + 1
    if numpy.arccos(cos)*180/numpy.pi < 30:
        count2 = count2 + 1
        
count1 = count1/100
count2 = count2/100

print('Пропорция векторов, для которых угол(v,q) < 90 градусов=',count1)
print('Пропорция векторов, для которых угол(v,q) < 30 градусов=',count2)
```

```python
headers = ['item', 'country', 'year', 'sales']
path_csv = f'./sales.csv'
with open(path_csv,'a+', newline='') as csvfile:
    spawnrow = csv.DictWriter(csvfile, fieldnames=headers)
    spawnrow.writeheader()
with open('sales.json') as f:
    tmp = json.load(f)
with open(path_csv, 'a+', newline='') as csvfile:
    spawnrow = csv.DictWriter(csvfile, fieldnames=headers)
    for item in tmp:
        countrys = item["sales_by_country"]
        for country in countrys:
            years = countrys[country]
            for year in years:
                spawnrow.writerow(
                    {'item': f'{item["item"]}', 'country': f'{country}',
                        'year': f'{year}', 'sales': f'{years[year]}'})
```
